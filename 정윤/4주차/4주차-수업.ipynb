### 4주차 과제
    어쩌고 저쩌고 쌀라쌀라

```python
x_data = [2, 4, 6, 8]
y_data = [81, 93, 91, 97]
```

하이퍼 파라미터 선언 : 기울기, 절편, 학습률, 에포크 설정

```python
# 기울기
w = 0
# Y절편
b = 0
# 러닝레이트
lr = 0.0001
# 학습 반복 횟수
epochs = 10000000
```

numpy를 통한 모델 학습 : 경사하강법(Gradient descent 사용)
- w에 대한 MSE, b에 대한 MSE를 구함 (MSE라는 함수로 구현)
- MSE의 기울기를 구함
- 구한 기울기가 0인 지점에서 최소값을 가지므로 경사하강법으로 그때의 w값, b값을 구한다.
- 

```python
import numpy as np

x_data = np.array(x_data)
y_data = np.array(y_data)
print(x_data)
print(y_data)

for i in range(epochs):
  y_pred = w * x_data + b   # w, d 값으로 나온 예상값
  error = y_data - y_pred   # 실제 y값과 예상값의 차이

  w_diff = -(1 / len(x_data)) * sum(x_data * error)   # 평균 제곱 오차를 a로 미분한 값
  b_diff = -(1 / len(x_data)) * sum(y_data - y_pred)  # 평균 제곱 오차를 b로 미분한 값

  w = w - lr * w_diff   # 학습률 * 미분 결과 후 기존 a 값 업데이트
  b = b - lr * b_diff   # 학습률 * 미분 결과 후 기존 b 값 업데이트
  print(f'w값 업데이트 : {w}')
  print(f'b값 업데이트 : {b}')

```
